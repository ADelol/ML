## 2. Inpainting

```python
from inpainting import read_im,show_im,save_img,get_patch,patch2vector,vector2patch
from inpainting import noise,delete_rect,get_patch_miss,get_patch_miss_ordered,get_dict,get_patch_similar,set_patch
from inpainting import random_fill_img,simple_fill_img,better_fill_img

COLOR_MISS=-100
#h_patch>=step pour avoir des redondants
STEP=10
H_PATCH=12
```

### I.Test "Lena"

Image originale:

```python
lena_hsv=read_im("./lena.jpg")
show_im(lena_hsv)
```

Ajouter des bruits à l'image:

```python
lena_noise=noise(lena_hsv,0.005)
show_im(lena_noise)
```

Eliminer les bruits:

```python
res_noise=simple_fill_img(lena_noise)
show_im(res_noise)
```

<br/>Ajouter un rectangle "noir" dans l'image originale:

```python
lena_rect=delete_rect(lena_hsv,430,360,50,50)
show_im(lena_rect)
```

Réparer l'image en ordre aléatoire:

```python
res_rect_random=random_fill_img(lena_rect)
show_im(res_rect_random)
```

Réparer l'image en ordre gauche->droite, haut->bas:

```python
res_rect_simple=simple_fill_img(lena_rect)
show_im(res_rect_simple)
```

Réparer l'image en ordre amélioré:

```python
res_rect_better=better_fill_img(lena_rect)
show_im(res_rect_better)
```
